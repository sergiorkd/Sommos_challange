```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Tratamento dos dados

```python
df_essay = pd.read_csv('essay-br.csv')
df_prompt = pd.read_csv('prompts.csv')
```

```python
import ast
# Renomear coluna prompt para corresponder com o nome da tabela prompt
df_essay.rename(columns={'prompt':'id'}, inplace=True)
# Juntar as duas tabelas em uma
df_essay = df_essay.merge(df_prompt, on='id', how='left')
# Separar a coluna competence em 5 categorias
df_essay['competence'] = df_essay['competence'].apply(ast.literal_eval)
competence_columns = df_essay['competence'].apply(pd.Series)
competence_columns.columns = [
    'Língua Portuguesa',
    'Tema e Gênero',
    'Coerência',
    'Coesão',
    'Proposta de Intervenção'
]
# Concatenando as tabelas lado a lado
df_essay = pd.concat([df_essay, competence_columns], axis=1)
```

```python
# Função para calcular a quantidade de palavras únicas
def voc_complexity(essay):
    words = essay.split()
    unique_words = set(words)
    return len(unique_words)

# Aplicar a função e criar uma nova coluna com a quantidade de palavras únicas
df_essay['unique_words'] = df_essay['essay'].apply(voc_complexity)
```

```python
# Filtrar as notas diferentes de zero
df_notas_positivas = df_essay[df_essay['score'] > 0]

# Calcular a média das notas por tema
tema_score_medio = df_notas_positivas.groupby('id')['score'].mean()

# Identificar o tema com a menor nota média (diferente de zero)
tema_menor_nota = tema_score_medio.idxmin()
nota_menor = tema_score_medio.min()

# Identificar o(s) tema(s) com a maior nota média
nota_maior = tema_score_medio.max()
temas_maior_nota = tema_score_medio[tema_score_medio == nota_maior].index.tolist()

# Contar a frequência de notas zero por tema
tema_zero_count = df_essay[df_essay['score'] == 0].groupby('id').size()

# Identificar o tema com a maior concentração de notas zero
tema_maior_zeros = tema_zero_count.idxmax()
maior_zeros = tema_zero_count.max()

# Resultados
print(f"Id do prompt com a menor nota média (diferente de zero): {tema_menor_nota} - Nota média: {nota_menor}")
print(f"Id do prompt com a maior nota média: {temas_maior_nota} - Nota média: {nota_maior:.2f}")
print(f"Id do prompt com a maior concentração de notas zero: {tema_maior_zeros} - Quantidade de notas zero: {maior_zeros}")
```

```python
# Selecionar apenas as colunas de interesse: as 5 competências e o score
df_competence = df_essay.iloc[:, [6,7,8,9,10,-1,-2,4]]

# Calcular a matriz de correlação
correlation_matrix = df_competence.corr()

# Plotar a matriz de correlação com um mapa de calor
plt.figure(figsize=(5, 3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlação entre Competências e Score')
plt.show()
```

Como era de se esperar, pode ser observado pela matriz de correlação, que as notas individuais das competências tem forte correlação linear positiva com a nota geral.  
No entando é interessante observar que a correlação entre a nota de coesão e o score é mais elevada, o que pode indicar que canditados que escrevem textos mais coesos tendem a ter notas maiores.  
Também pode ser observado que a relação linear entre a nota de lingua portuguesa e a nota geral é a menor, indicando que a maioria dos canditados apresenta um nível parecido de acerto nesta categoria, como pode ser observado no histograma das notas.

```python
from scipy.stats import f_oneway

# Filtrar o DataFrame para as colunas desejadas
df_anova = df_essay.iloc[:, [6, 7, 8, 9, 10, -1, -2, 4]]

# Separar a variável target
target = df_anova['score']

# Criar uma lista para armazenar os resultados do ANOVA
anova_results = {}

# Iterar sobre as colunas de interesse (exceto a variável target)
for column in df_anova.columns[:-1]:
    # Aplicar o teste ANOVA
    groups = [df_anova[df_anova['score'] == score][column].dropna() for score in target.unique()]
    f_stat, p_value = f_oneway(*groups)
    anova_results[column] = (f_stat, p_value)

# Converter os resultados em um DataFrame para fácil visualização
anova_results_df = pd.DataFrame(anova_results, index=['F-Statistic', 'p-value']).T

# Exibir os resultados do ANOVA
print(anova_results_df)
```

A análise do teste ANOVA revelou que todos os p-valores obtidos foram extremamente baixos, indicando que rejeitamos a hipótese nula para todas as variáveis examinadas.   
Isso sugere uma diferença significativa nas notas dos alunos em relação a cada uma das variáveis, confirmando seu impacto real nas avaliações, como já se era de esperar.   
A variável "Coesão" destacou-se com o maior F-Statistic (1002.83), sugerindo que a habilidade dos alunos em conectar ideias e manter a fluidez do texto é um fator determinante na avaliação de suas redações, como idicado anteriormente pelo alto índice de correlação.  


Separarar o conjunto em dois grupos para testar a influencia da nota da competência Coesão nos resultados gerais. Separar o conjunto entre condidatos que tiraram acima de 125 em coesão e candidatos que tiraram abaixo de 125 e fazer um teste T de Student para testar a hipótese:  
Hipótese nula (H₀): A média da nota total para o grupo com "Coesão" alta é igual à média do grupo com "Coesão" baixa, indicando que a competência "Coesão" pode não impactar significativamente o score total.  
Hipótese alternativa (H₁): A média da nota total para o grupo com "Coesão" alta é significativamente diferente (ou maior) da média do grupo com "Coesão" baixa, sugerindo que, na prática, "Coesão" impacta as notas totais de maneira significativa.

```python
from scipy.stats import ttest_ind

# Separando os grupos com base na nota de Coesão
grupo_coesao_alta = df_essay[df_essay.iloc[:, 9] > 125]["score"]
grupo_coesao_baixa = df_essay[df_essay.iloc[:, 9] <= 125]["score"]

# Realizando o teste T de Student
t_stat, p_value = ttest_ind(grupo_coesao_alta, grupo_coesao_baixa)

# Exibindo os resultados
print("Estatística t:", t_stat)
print("p-valor:", p_value)

# Avaliação do resultado
if p_value < 0.05:
    print("Rejeitamos a hipótese nula. A competência 'Coesão' impacta significativamente o score total.")
else:
    print("Não rejeitamos a hipótese nula. A competência 'Coesão' pode não impactar significativamente o score total.")
```

A competência 'Coesão' impacta significativamente o score total, como já era de se esperar e como foi mostrado no teste ANOVA e pela alta correlação entre a variável score e a nota em coesão.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Remover colunas de índices 0, 1, 2, 3 e 5
df_numeric = df_essay.drop(df_essay.columns[[0, 1, 2, 3, 5]], axis=1)

# Escalar os dados para normalização
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numeric)

# Determinar o número de clusters usando o método do cotovelo
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Plotar o gráfico do método do cotovelo
plt.figure(figsize=(6, 3))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inércia')
plt.title('Método do Cotovelo para Determinar o Número de Clusters')
plt.show()

# Escolhendo o número ideal de clusters (suponha k=3 após análise do gráfico)
kmeans = KMeans(n_clusters=3, random_state=42)
df_essay['cluster'] = kmeans.fit_predict(df_scaled)
```

```python
# Calcular as médias das notas individuais e do número de palavras para cada cluster
# Supondo que "nota_total", "coesao" e "numero_palavras" sejam as colunas relevantes
cluster_analysis = df_essay.groupby('cluster').agg(
    media_nota_total=('score', 'mean'),
    media_coesao=('Coesão', 'mean'),
    media_numero_palavras=('num_words', 'mean')
).reset_index()

print("Média das características por cluster:")
print(cluster_analysis)

# Visualizar as médias com um gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(data=cluster_analysis.melt(id_vars='cluster'), x='variable', y='value', hue='cluster')
plt.title('Média das Notas e Número de Palavras por Cluster')
plt.xlabel('Características')
plt.ylabel('Média')
plt.legend(title='Cluster')
plt.xticks(rotation=45)
plt.show()

# Boxplots para visualizar a distribuição das notas por cluster
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_essay, x='cluster', y='score')
plt.title('Distribuição das Notas Totais por Cluster')
plt.xlabel('Cluster')
plt.ylabel('Nota Total')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_essay, x='cluster', y='num_words')
plt.title('Distribuição do Número de Palavras por Cluster')
plt.xlabel('Cluster')
plt.ylabel('Número de Palavras')
plt.show()
```
